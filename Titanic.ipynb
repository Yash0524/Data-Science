```python
df.drop('Cabin', axis=1, inplace=True)
```

# **Segregating Targets and Features.**

```python
A = df.drop(columns = ['PassengerId', 'Name', 'Ticket', 'Survived'], axis = 1)
B = df['Survived']
```

```python
print(A)
```

```python
print(B)
```

```python
from sklearn.model_selection import train_test_split

A_tr, A_ts, B_tr, B_ts = train_test_split(A, B, test_size = 0.2, random_state = 2)
```

```python
print(A_tr, A_ts)
```

```python
print(A_tr.shape, A_ts.shape)
```

```python
print(B_tr, B_ts)
```

```python
print(B_tr.shape, B_ts.shape)
```

# **Importing Libraries for Model Training**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# **Model Training**

```python
m = LogisticRegression()
```

```python
m.fit(A_tr, B_tr)
```

```python
A_tr_pred = m.predict(A_tr)
print(A_tr_pred)
```

```python
A_ts_pred = m.predict(A_ts)
print(A_ts_pred)
```

```python
tr_acc = accuracy_score(B_tr, A_tr_pred)
print("Accuracy of Training data : " , tr_acc)
```

```python
ts_acc = accuracy_score(B_ts, A_ts_pred)
print("Accuracy of Testing data : ", ts_acc)
```
