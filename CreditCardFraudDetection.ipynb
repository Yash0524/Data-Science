```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score
```

```python
df = df.dropna()
x = df.drop('Class', axis = 1)
y = df.Class

x_tr, x_ts, y_tr, y_ts = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 2)
```

```python
print(x_tr.shape, x_ts.shape, y_tr.shape, y_ts.shape)
```

```python
m = LogisticRegression()
m.fit(x_tr, y_tr)
```

```python
x_tr_pred = m.predict(x_tr)
tr_acc = accuracy_score(x_tr_pred, y_tr)
print("Accuracy on Training data : ", tr_acc)
```

```python
x_ts_pred = m.predict(x_ts)
ts_acc = accuracy_score(x_ts_pred, y_ts)
print("Accuracy on Testing data : ", ts_acc)
```

```python
print("Precision score on Testing data : ", precision_score(y_ts, x_ts_pred) * 100)
```

```python
sample = m.predict(x_ts)
print("Legal" if (sample[0] == 0) else "Fraud")
```
